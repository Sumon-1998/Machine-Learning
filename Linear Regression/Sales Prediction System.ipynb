```python
import numpy as np
import pandas as pd
```

# Reading the data set

```python
data=pd.read_csv("/home/sumon/Data Science Note/Advertising.csv",index_col=0)
data
```

# Visualizing the different columns of the dataset

```python
import seaborn as sb
sb.pairplot(data,x_vars=['TV','Radio','Newspaper'],y_vars=['Sales'],aspect=0.7,height=7,kind='reg')
```

```python
x=data[['TV','Radio','Newspaper']]
y=data['Sales']
```

# Split the data set into training data and testing data

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
x_train
```

```python
y_train
```

```python
x_test
```

```python
y_test
```

# Training our model

```python
from sklearn.linear_model import LinearRegression
linreg=LinearRegression()
linreg.fit(x_train,y_train)
```

# Interpreting model coefficients

```python
print(linreg.coef_)
print(linreg.intercept_)
```

# Making Predictions

```python
y_pred=linreg.predict(x_test)
y_pred
```

```python
y_test
```

```python
comparison_data=pd.DataFrame()
comparison_data['y_test']=y_test
comparison_data['y_predict']=y_pred
comparison_data
```

# Evaluation metrics of model

```python
from sklearn import metrics
```

```python
#mean absolute error
m1=metrics.mean_absolute_error(y_test,y_pred)
m1
```

```python
#mean squared error
m2=metrics.mean_squared_error(y_test,y_pred)
m2
```

```python
#root mean squared error
m3=np.sqrt(metrics.mean_squared_error(y_test,y_pred))
m3
```
